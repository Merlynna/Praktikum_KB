## Electric Vehicle Population Data Set

```python
import pandas as pd
dataFrame = pd.read_csv('Electric_Vehicle_Population_Data.csv')
dataFrame.head()
```

### Link DataSet : https://drive.google.com/file/d/1UvsonUWSBUJ5Rn4Ofb7VF_p8zOHgewwl/view?usp=drive_link

## Data Cleaning

```python
# Periksa jika ada salah satu dari atribut yang nilainya tidak ada
dataFrame.isna().sum()
```

## Kosongkan Data

```python
# Menandakan apakah dalam atribut ini ada nilai yang Kosong
print("Jumlah yang memiliki nilai Null : ")
dataresult = dataFrame['Electric Range'].isna().sum()
print(dataresult)
```

## Clear Null

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'mean')
dataFrame["Electric Range"] = imputer.fit_transform(dataFrame[["Electric Range"]])
dataFrame
```

```python
print("Berikut ini adalah hasil setelah melakukan Clear Null : ")
dataFrame.isna().sum()
```

```python
# Menunjukkan bahwa data Null pada atribut sudah tidak ada lagi
dataFrame['Electric Range'].isna().sum()
```

## Menangani Nilai Duplikasi

```python
# Menunjukkan bahwa nilai duplikasi pada DataFrame tidak ada
dataFrame.duplicated().sum()
```

## Melakukan Duplikasi

```python
# Menunjukkan bahwa data pada Baris indeks [0] dan indeks [1] terduplikasi
dataFrame.iloc[1,:] = dataFrame.iloc[0,:]
dataFrame
```

## Hilangkan Duplikasi

```python
# Menunjukkan bahwa nilai duplikasi pada Baris indeks [0] dan indeks [1] kembali seperti semula
dataFrame.drop_duplicates(inplace = True)
dataFrame.head()
```

## Proses Normalisasi

```python
# Untuk mengubah nilai numerik pada data
from sklearn.preprocessing import MinMaxScaler
normal = MinMaxScaler()
dataFrame["Electric Range"] = normal.fit_transform(dataFrame[["Electric Range"]])
dataFrame.head()
```

## Pembuktian Max dan Min

```python
# Menunjukkan bahwa Maksimum Electric Range adalah sebesar 1
dataFrame["Electric Range"].max()
```

```python
# Menunjukkan bahwa Maksimum Electric Range adalah sebesar 0 atau null
dataFrame["Electric Range"].min()
```

## Proses Standarisasi

```python
# Untuk mengubah nilai numerik pada data
from sklearn.preprocessing import StandardScaler
standard = StandardScaler()
dataFrame["Postal Code"] = standard.fit_transform(dataFrame[["Postal Code"]])
dataFrame.head(20)
```

## Pembuktian Mean dan Standard Deviasi

```python
# Menunjukkan bahwa Mean pada atribut ini kurang dari 0 atau mendekati 0
dataFrame['Postal Code'].mean()
```

```python
# Menunjukkan bahwa Standard Deviasi sama dengan 1
dataFrame['Postal Code'].std()
```

## OneHot Encoder

```python
from sklearn.preprocessing import OneHotEncoder
# OneHot Encoder
OneHot = OneHotEncoder(sparse_output=False)
dfEnc = OneHot.fit_transform(dataFrame[['City']])

#Ubah Ke Dataframe
dfEnc = pd.DataFrame(
    dfEnc
)
```

```python
dfEnc.head()
```

```python
dataFrame['City'].value_counts()
```

```python
dataFrame["Decatur"] = dfEnc[0]
dataFrame["Bellevue"] = dfEnc[1]
dataFrame["Redmond"] = dfEnc[2]
```

```python
# Menampilkan hasil keseluruhan data
dataFrame
```

## Data Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
# Features / Atribute
X = dataFrame.drop("County", axis = 1)

# Target
y = dataFrame['County']
```

```python
# Menunjukkan bahwa dalam setiap dataFrame Jumlah Train_size = 70% sedangkan Test_size = 30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
print("Jumlah X_Train : ", X_train.shape)
print("Jumlah X_Test  : ", X_test.shape)
print("Jumlah y_Train : ", y_train.shape)
print("Jumlah y_Test  : ", y_test.shape)
```
