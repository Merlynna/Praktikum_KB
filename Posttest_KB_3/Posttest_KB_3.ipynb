## Tampilkan Data dari CSV

```python
dataFrame = pd.read_csv("Electric_Vehicle_Population_Data.csv")
dataFrame

dataFrame.head(10)
```

## Tampilan Grafik Pie (Pie Chart)

```python
# Tampilkan Grafik Pie jika menampilkan 5 data terawal
dataPie = dataFrame.Make.value_counts().head(5)
plt.pie(dataPie, labels = dataPie.index, autopct = "%1.1f%%")
plt.title("Tampilan Pie Chart Menggunakan Atribut Make")
plt.xlabel("Nama City")
plt.ylabel("Persentasi Jumlah")
plt.show()
```

```python
# Tampilkan Grafik Pie jika menampilkan 4 data terawal
dataPie = dataFrame.Make.value_counts().head(4)
plt.pie(dataPie, labels = dataPie.index, autopct = "%1.1f%%")
plt.title("Tampilan Pie Chart Menggunakan Atribut Make")
plt.xlabel("Nama City")
plt.ylabel("Persentasi Jumlah")
plt.show()
```

```python
# Tampilkan Grafik Pie jika menampilkan 4 data terakhir
dataPie = dataFrame.Make.value_counts().tail(4)
plt.pie(dataPie, labels = dataPie.index, autopct = "%1.1f%%")
plt.title("Tampilan Pie Chart Menggunakan Atribut Make")
plt.xlabel("Nama City")
plt.ylabel("Persentasi Jumlah")
plt.show()
```

```python
# Tampilkan Grafik Pie jika menampilkan 4 data terakhir
dataPie = dataFrame.City.value_counts().tail(4)
plt.pie(dataPie, labels = dataPie.index, autopct = "%1.1f%%")
plt.title("Tampilan Pie Chart Menggunakan Atribut City")
plt.xlabel("Nama City")
plt.ylabel("Persentasi Jumlah")
plt.show()
```

## Tampilan Grafik Histogram

```python
# Grafik Histogram Model Year
plt.hist(dataFrame['Model Year'], bins = 15, color = 'pink')
plt.title('Histogram Jumlah Model Year')
plt.ylabel('Frekuensi Digunakan')
plt.xlabel('Tahun Model')
plt.show()
```

## Tampilan Grafik Batang

```python
plt.figure(figsize=(10,5))
model = ['2019', '2022', '2020', '2020', '2013', '2018', '2020', '2021', '2019', '2018']
county = ['Thurston', 'Island', 'Shohomish', 'King', 'Shohomish', 'Chelan', 'Shohomish', 'Shohomish', 'Thurston', 'Thurston']

plt.bar(model, county, color = "orange")
plt.title("Tampilan Grafik Bar antara Tahun Model dengan Country")
plt.xlabel("Tahun Model")
plt.ylabel("Country")
plt.show()
plt.close()
```
